## Q 10. Plot the distribution plot for the variable 'MatchesPlayed'. Write detailed observations from the plot.

Here we have plotted the distribution of Matches played in different worldcup Fixture . Here We can see from the plot that in beggining the matches played was very less but in some recent world cup fixtures the qualifed and teams and matches played increased.

```python
sns.set(style="whitegrid")
sns.histplot(world_cups_df['MatchesPlayed'], kde=True, bins=10)
plt.title('Distribution of Matches Played in FIFA World Cups')
plt.xlabel('Matches Played')
plt.ylabel('Frequency')
plt.show()
```

## Q 11. Which country has won the world cup maximum times?


```python
# Here we have used the groupby function and I had used group by on Winner column to get the list of the teams who won most of the times
# After getting that list we can simply sort that list using Ascending as false. After that if we print the index 0 then it will print the maximum times world cup winner

wins_by = world_cups_df.groupby('Winner').size().sort_values(ascending= False)
print("Country with maximum wins:", wins_by.index[0])
print(wins_by)
```

## Q12.
### 12.1 What is the mean of the variable 'Qualified teams'?
### 12.2 What is the median of the variable 'Qualified teams'?
### 12.3 What is the mode of the variable 'Qualified teams'?


* The mean (average) of a data set is found by adding all numbers in the data set and then dividing by the number of values in the set.
* The median is the middle value when a data set is ordered from least to greatest.
* The mode is the number that occurs most often in a data set.

```python
mean_qualified_teams = world_cups_df['QualifiedTeams'].mean()

print("Mean of the variable 'Qualified teams':", mean_qualified_teams)

median_qualified_teams = world_cups_df['QualifiedTeams'].median()

print("Median of the variable 'Qualified teams':", median_qualified_teams)

mode_qualified_teams = world_cups_df['QualifiedTeams'].mode()[0]

print("Mode of the variable 'Qualified teams':", mode_qualified_teams)

world_cups_df.head(20)
```

## Q13. How many countries are above the mean level of 'Qualified Teams'?

```python
mean_qualified = world_cups_df['QualifiedTeams'].mean()
above_mean_countries = world_cups_df[world_cups_df['QualifiedTeams'] > mean_qualified]
print("Number of countries above the mean level of 'Qualified Teams':", len(above_mean_countries))
```

## Q14. What is the median of variables 'GoalsScored' & 'MatchesPlayed'?

```python
median_goals_scored, median_matches_played = world_cups_df[['GoalsScored', 'MatchesPlayed']].median()

print("Median of the variable 'GoalsScored':", median_goals_scored)
print("Median of the variable 'MatchesPlayed':", median_matches_played)
```

## Q15. Which country scored the minimum number of goals?

```python
min_goals_scored_country = world_cups_df[world_cups_df['GoalsScored'] == world_cups_df['GoalsScored'].min()]['Country'].iloc[0]
print("Country with the minimum number of goals:", min_goals_scored_country)
```

## Q17. Plot the scatterplot for variables 'Country' & 'year'.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=world_cups_df, x='Year', y='Country', hue = 'Country', palette='Set1')
plt.xlabel('Year')
plt.ylabel('Country')
plt.title('Scatterplot of Country vs. Year')
plt.show()
```

## Q18. Plot a histogram for the variable 'Winner' to understand the number of times a country won the world cup between 1930 to 2014.

```python
filtered_df = world_cups_df[(world_cups_df['Year'] >= 1930) & (world_cups_df['Year'] <= 2014)]
plt.figure(figsize=(10, 6))
sns.histplot(filtered_df['Winner'], color='skyblue', edgecolor='black')
plt.title('Number of World Cup Wins by Country (1930-2014)')
plt.xlabel('Country')
plt.ylabel('Number of Wins')
plt.show()
```

## Q 19. Show boxplot and calculate the interquartile range for the variable 'GoalsScored'

```python
goals_scored = world_cups_df['GoalsScored']


plt.figure(figsize=(10, 6))
plt.boxplot(goals_scored, vert=False)
plt.title('Boxplot of Goals Scored in World Cup')
plt.xlabel('Goals Scored')
plt.show()

Q1 = goals_scored.quantile(0.25)
Q3 = goals_scored.quantile(0.75)
IQR = Q3 - Q1

Q1, Q3, IQR

#The interquartile range (IQR) for the 'GoalsScored' variable is calculated as follows:
#First Quartile (Q1): 89.0
#Third Quartile (Q3): 145.25
#Interquartile Range (IQR): 56.25 ​
```

## Q 20. Find and visualize the correlation relation among numeric variables

```python
numeric_columns = world_cups_df.select_dtypes(include=['int64', 'float64'])

# Compute the correlation matrix
correlation_matrix = numeric_columns.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Numeric Variables')
plt.show()
```
